## Attack and defense by primary type

### Find the best pokemon (having the largest basic attack) in every primary type (type1)
### Plot the top 10 types having the strongest pokemon (type + pokemon name)
### Do the same for defense
### Bonus: plot the attacka nd ddefense for every pokemon and color by primary type

```python
rcParams['figure.figsize'] = 15, 10
pokemon["name_type1"]=pokemon["type1"]+": "+pokemon["name"]
pokemon[["type1","attack","name_type1"]].groupby(pokemon["type1"]).max().set_index("name_type1").sort_values(by=['attack'], ascending=False).plot(kind="barh")
```

```python
pokemon[["type1","name_type1","defense"]].groupby(pokemon["type1"]).max().set_index("name_type1").sort_values(by=['defense'], ascending=False).plot(kind="bar")
```

```python
groups = pokemon[["attack","defense","name","type1"]].groupby('type1')
rcParams['figure.figsize'] = 10, 10
# Plot
fig, ax = plt.subplots()
ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group.attack, group.defense, marker='o', linestyle='', ms=8, label=name)
ax.legend()
plt.xlabel('basic attack')
plt.ylabel('basic defense')
plt.show()
```
